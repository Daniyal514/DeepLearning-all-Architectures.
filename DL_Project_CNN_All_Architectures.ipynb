```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, BatchNormalization, AveragePooling2D, GlobalAveragePooling2D, concatenate,Add,MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Nadam
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

```python
tr=pd.read_csv('/content/drive/MyDrive/train.csv')
print(np.shape(tr))
```

```python
def load_data():
    x_train, x_val, y_train, y_val = train_test_split(
        tr.drop('label', axis=1).values, tr['label'].values,
        random_state=31, train_size=0.7, shuffle=True, stratify=tr['label'].values
    )

    print("New Training data shape:", x_train.shape)
    print("New Training labels shape:", y_train.shape)
    print("New Validation data shape:", x_val.shape)
    print("New Validation labels shape:", y_val.shape)

    return x_train, x_val, y_train, y_val
```

```python
x_train,x_val,y_train,y_val=load_data()
mean=np.mean(x_train,axis=0)
x_train=(x_train-mean)/255
x_val=(x_val-mean)/255


y_train = to_categorical(y_train, 10)  # 10 classes in CIFAR-10
y_val =  to_categorical(y_val, 10)  # 10 classes in CIFAR-10
```

```python

x_train = x_train.reshape(-1, 32, 32, 3)
x_val = x_val.reshape(-1, 32, 32, 3)
```

```python
#DenseNET
def dense_block(x, num_filters, num_layers):
    for _ in range(num_layers):
        y = BatchNormalization()(x)
        y = Conv2D(num_filters, kernel_size=(3, 3), padding='same', activation='relu')(y)
        x = concatenate([x, y])
    return x

def transition_layer(x, num_filters):
    x = BatchNormalization()(x)
    x = Conv2D(num_filters, kernel_size=(1, 1), activation='relu')(x)
    x = AveragePooling2D(pool_size=(2, 2), strides=(2, 2))(x)
    return x

input_shape = (32, 32, 3)
num_classes = 10

input_tensor = Input(shape=input_shape)
x = Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu')(input_tensor)

# Create Dense Blocks with Transition Layers
x = dense_block(x, 32, 6)
x = transition_layer(x, 64)
x = dense_block(x, 32, 12)
x = transition_layer(x, 128)
x = dense_block(x, 32, 24)
x = transition_layer(x, 256)
x = dense_block(x, 32, 16)

x = BatchNormalization()(x)
x = GlobalAveragePooling2D()(x)
x = Flatten()(x)
x = Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.00001))(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu', kernel_regularizer=regularizers.l2(0.00001))(x)
x = Dropout(0.5)(x)
output_tensor = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=input_tensor, outputs=output_tensor)

model.compile(optimizer=Nadam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Reduce learning rate when a metric has stopped improving
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.00001)

# Print the model summary
model.summary()
```

```python
# Fit the model
history = model.fit(x_train, y_train, batch_size=128, epochs=20, validation_data=(x_val, y_val), callbacks=[reduce_lr])
```

```python
# Calculate validation accuracy
val_loss, val_accuracy = model.evaluate(x_val, y_val)
print("Validation Accuracy:", val_accuracy)

# Load test data
ts = pd.read_csv('/content/drive/MyDrive/test.csv')
print(np.shape(ts))

test = ts.iloc[:, 1:].values
index = ts.iloc[:, 0]

# Normalize test data
test = (test - mean) / 255.0
test = test.reshape(-1, 32, 32, 3)  # Reshape to fit LeNet input requirements

```

```python


# Predict test data
Pred_labels = model.predict(test)

# Convert predictions from probabilities to class labels
predicted_classes = np.argmax(Pred_labels, axis=1)

# Create a DataFrame for submission
df = pd.DataFrame({'id': index, 'label': predicted_classes})

# Save the DataFrame to a CSV file
df.to_csv('submission_roll_no.csv', index=False)
```
